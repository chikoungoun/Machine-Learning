 separate the array elements into different bins 

```python
import numpy as np
import pandas as pd
```

```python
df= pd.DataFrame({'number': np.random.randint(1, 100, 10)})
```

```python
df
```

```python
# classify the data into bins range
df['bins'] = pd.cut(df['number'], bins=[1,20,40,60,80,100])
df
```

```python
# naming each label
df['labels'] = pd.cut(df['number'], bins=[1,20,40,60,80,100],labels=[1,2,3,4,5])
df
```
